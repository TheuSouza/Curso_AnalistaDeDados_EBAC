```python
import seaborn as sns
import matplotlib.pyplot as plt


flights = sns.load_dataset('flights')
flights.head()
```

```python
passengers_december = flights.query('month == "Dec"').reset_index()
del passengers_december['index']
passengers_december.head()
```

```python
plt.figure(figsize=(12, 6))
grafico = sns.lineplot(data=passengers_december, x='year', y='passengers', hue='month', palette='rocket')
grafico.set(title='Nmero de passageiros em dezembro por ano', xlabel='Ano', ylabel='Passageiros')
grafico.get_legend().set_title('Ms')
plt.grid(linestyle='--', alpha=0.3)
```

```python
grafico.figure.savefig('passengers_december.png', bbox_inches='tight')
```

```python
passengers_mes = flights.query('1949 <=  year <= 1959')
passengers_mes.tail()
```

```python
plt.figure(figsize=(12, 6))
grafico2 = sns.lineplot(data=passengers_mes, x='year', y='passengers', hue='month', palette='rocket', size='month')
grafico2.set(title='Nmero de passageiros por ms entre 1949 e 1959', xlabel='Ano', ylabel='Passageiros')
grafico2.get_legend().set_title('Ms')
plt.grid(linestyle='--', alpha=0.3)
```

```python
grafico2.figure.savefig('passengers_month.png', bbox_inches='tight')
```
